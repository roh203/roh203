```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('/content/drive/MyDrive/mill(1).csv')
```

```python
df_x = df.iloc[:,:-1]
df_y = df.iloc[:,-1]
```

```python
df_x
```

```python
df_y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ns = MinMaxScaler()
ns.fit(df_x)
```

```python
df_x = ns.transform(df_x)
data = pd.DataFrame(df_x)
data.columns=['YV','FM','FM','TSSFMWB','TSSCWAD','TRSMRP','WFMpH','WFH']
df_x =data
```

```python
df_x.head()
```

```python
df_x.to_csv('normalized_data.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y, test_size = .2, random_state = 103)
```

```python
x_train.to_csv('train_data.csv')
```

```python
y_train.to_csv('y_train.csv')
```

```python
x_test.to_csv('x_test_data.csv')
```

```python
y_test.to_csv('y_test_data.csv')
```

```python
model = Sequential()
model.add(Dense(8, input_dim=8, activation='relu'))  # # input layer
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
import tensorflow as tf
tf.keras.utils .plot_model(model,to_file='simple.png',show_shapes=True)
```

```python
model.save
```

```python
from sklearn.metrics import r2_score
```

```python
model.compile(loss= 'mse', optimizer='adam', metrics = ['accuracy'])
history= model.fit(x_train, y_train, epochs=250, batch_size=5, verbose=5, validation_split=0.2)
plt.figure(figsize=(8, 8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
fig = plt.show()
y_pred = model.predict(x_test)
ytrain_pred = model.predict(x_train)
print(r2_score(y_train,ytrain_pred))
print(r2_score(y_test,y_pred))
```
